```python
#Drop unwanted variables from data
churn_regression = df.columns
X_vals = df.loc[:,df.columns !='Churn_yes']
y_vals = df['Churn_Yes']

churn_regression = X_vals.columns
drop_cols = ['Churn_No','City','County','State','Interaction','DeviceProtection_No',
             'Customer_id','TechSupport_No','OnlineBackup_No','OnlineSecurity_No',
            'Phone_No','Port_modem_No','Multiple']
X_vals = df.drop(drop_cols, axis=1)
X_vals = X_vals.drop('Job', axis=1)
X_vals = X_vals.drop('UID', axis=1)
X_vals = X_vals.drop('Churn_Yes', axis=1)
X_vals = X_vals.drop('TimeZone', axis=1)
X_vals = X_vals.drop('Area', axis=1)
X_vals = X_vals.drop('Gender', axis=1)
X_vals = X_vals.drop('Techie', axis=1)
X_vals = X_vals.drop('Tablet', axis=1)
X_vals = X_vals.drop('PaperlessBilling', axis=1)
X_vals = X_vals.drop('StreamingMovies_No', axis=1)
X_vals = X_vals.drop('StreamingTV_No', axis=1)
X_vals = X_vals.drop('PaymentMethod_Bank Transfer(automatic)',axis=1)
X_vals = X_vals.drop('Marital_Divorced', axis=1)
X_vals = X_vals.drop('InternetService_None',axis=1)

X_vals.info()
```

```python
#Import package to split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_vals, y_vals, test_size=0.80, random_state=234)
```

```python
#Import package to perform logisitc regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(X_train, y_train)
```

```python
#Import package to display logistic regression output
import statsmodels.api as sm
model_stats = sm.Logit(y_train, X_train).fit()
model_stats.summary()
```

```python
#Dispaly coefficents of output
model.coef_
```

```python
#Perform predictions
model.predict(X_test)
```

```python
#Get score of predictions
model.score(X_test,y_test)
```

```python
#Print accuracy and precision
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, model.predict(X_test)))
```

```python
#Confusion matrix of predictions 
cm = confusion_matrix(y_test, model.predict(X_test))

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='red')
plt.show()
```
